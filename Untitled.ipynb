```python
def get_compound_transactions_alchemy(wallet_address):
   
    try:
        url = f"https://eth-mainnet.g.alchemy.com/v2/{ALCHEMY_API_KEY}"
        payload = {
            "jsonrpc": "2.0",
            "method": "alchemy_getAssetTransfers",
            "params": [{
                "fromBlock": "0x0",
                "toBlock": "latest",
                "fromAddress": wallet_address,
                "toAddress": list(COMPOUND_V2_CONTRACTS),
                "excludeZeroValue": False,
                "category": ["external", "internal", "erc20", "erc721", "erc1155"]
            }],
            "id": 1
        }
        headers = {"accept": "application/json", "content-type": "application/json"}
        response = requests.post(url, json=payload, headers=headers)
        data = response.json()
        return data.get('result', {}).get('transfers', [])
    except Exception as e:
        print(f"Alchemy error for {wallet_address}: {str(e)}")
        return []
```

```python
def get_compound_transactions(wallet_address):
    
    try:
    
        params = {
            'module': 'account',
            'action': 'txlist',
            'address': wallet_address,
            'startblock': 0,
            'endblock': 'latest',
            'sort': 'asc',
            'apikey': ETHERSCAN_API_KEY
        }
        response = requests.get('https://api.etherscan.io/api', params=params)
        data = response.json()
        
        if data['status'] == '1':
            etherscan_txs = [tx for tx in data['result'] 
                           if tx['to'].lower() in COMPOUND_V2_CONTRACTS]
            if etherscan_txs:
                return etherscan_txs
        
        return get_compound_transactions_alchemy(wallet_address)
    except Exception as e:
        print(f"Error fetching txs for {wallet_address}: {str(e)}")
        return []
```

```python
def extract_features(wallet_address, transactions):
    """Extract risk features from transactions"""
    if not transactions:
        return None
    
    features = {
        'wallet_id': wallet_address,
        'total_txs': len(transactions),
        'days_active': 1,
        'borrow_count': 0,
        'repay_count': 0,
        'liquidation_count': 0,
        'flashloan_count': 0,
        'failed_txs': 0,
        'high_value_txs': 0,
    }
    
   
    for tx in transactions:
        # Count failed transactions
        if tx.get('isError') == '1' or tx.get('receipt_status') == '0':
            features['failed_txs'] += 1
            continue
            
     
        value = float(tx.get('value', 0)) / 1e18
        if value > 10:
            features['high_value_txs'] += 1
            
       
        input_data = tx.get('input', '')
        if len(input_data) >= 10:
            method = input_data[:10]
            if method == '0xc5ebeaec':  # borrow
                features['borrow_count'] += 1
            elif method == '0xe3a9db1a':  # liquidate
                features['liquidation_count'] += 1
            elif method == '0x5cffe9de':  # flashloan
                features['flashloan_count'] += 1
            elif method == '0x0e752702':  # repay
                features['repay_count'] += 1
    
 
    features['tx_failure_rate'] = (features['failed_txs'] / features['total_txs'] 
                                  if features['total_txs'] > 0 else 0)
    features['borrow_ratio'] = (features['borrow_count'] / (features['repay_count'] + 1))
        
    return features


def calculate_risk_score(features):
    """Calculate risk score (0-1000)"""
    if not features:
        return 500  
    
   
    weights = {
        'tx_failure_rate': 0.20,
        'liquidation_count': 0.25,
        'borrow_ratio': 0.25,
        'flashloan_count': 0.10,
        'high_value_txs': 0.10,
        'borrow_count': 0.10
    }
    
    normalized = {
        'tx_failure_rate': min(features.get('tx_failure_rate', 0), 1),
        'liquidation_count': min(features.get('liquidation_count', 0) / 5, 1),  # 0-5 range
        'borrow_ratio': min(features.get('borrow_ratio', 0), 10),  # Cap at 10
        'flashloan_count': min(features.get('flashloan_count', 0) / 3, 1),  # 0-3 range
        'high_value_txs': min(features.get('high_value_txs', 0) / 10, 1),  # 0-10 range
        'borrow_count': min(features.get('borrow_count', 0) / 20, 1)  # 0-20 range
    }
    
   
    score = sum(normalized[feature] * weight 
               for feature, weight in weights.items())
    
   
    final_score = min(int((score ** 0.7) * 1200), 1000)
    
  
    if features['total_txs'] > 0:
        final_score = max(final_score, 50)
        
    return final_score
```

```python
def process_wallets(wallet_list, sample_size=None):

    if sample_size:
        wallet_list = wallet_list[:sample_size]
    
    results = []
    
    for i, wallet in enumerate(wallet_list):
        try:
            print(f"Processing wallet {i+1}/{len(wallet_list)}: {wallet}")
            
            txs = get_compound_transactions(wallet)
            features = extract_features(wallet, txs)
            score = calculate_risk_score(features)
            
            results.append({
                'wallet_id': wallet,
                'score': score,
                'tx_count': features['total_txs'] if features else 0,
                'borrow_count': features.get('borrow_count', 0) if features else 0,
                'liquidation_count': features.get('liquidation_count', 0) if features else 0
            })
            
            time.sleep(0.5) 
            
        except Exception as e:
            print(f"Error processing {wallet}: {str(e)}")
            results.append({
                'wallet_id': wallet,
                'score': 500,
                'tx_count': 0,
                'borrow_count': 0,
                'liquidation_count': 0
            })
    
    return pd.DataFrame(results)
```

```python
# Run the Analysis
risk_scores_df = process_wallets(wallets)
```

```python
print("\nRisk Score Distribution:")
print(risk_scores_df['score'].describe())

plt.figure(figsize=(10, 6))
sns.histplot(risk_scores_df['score'], bins=20, kde=True)
plt.title('Wallet Risk Score Distribution')
plt.xlabel('Risk Score (0-1000)')
plt.ylabel('Number of Wallets')
plt.show()
```
